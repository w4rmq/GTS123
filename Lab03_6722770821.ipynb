```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-np.pi * 3, (np.pi * 3) + 0.1, 0.1)

f1 = np.sin(x)
f2 = np.sin(x+0.5)
f3 = np.sin(x+1)
f4 = np.sin(x+1.5)

plt.plot(x, f1, color="blue", linewidth=2, linestyle="dotted")
plt.plot(x, f2, color="green", linewidth=2, linestyle="--")
plt.plot(x, f3, color="orange", linewidth=2, linestyle="-.")
plt.plot(x, f4, color="pink", linewidth=2)
plt.show()
```

Question 9

```python
import math as m

cube_width = float(input("Enter the width of the cube: "))
container_width = float(input("Enter the width of the container: "))
container_height = float(input("Enter the height of the container: "))
container_depth = float(input("Enter the dept of the container: "))

num_width = m.floor(container_width / cube_width)
num_height = m.floor(container_height / cube_width)
num_depth = m.floor(container_depth / cube_width)

total = num_width * num_height * num_depth

print("The number of cubes that can fit into the container is %d." % total)
```
